### Top 5 Most Common Jobs

```python
import matplotlib.pyplot as plt

# Assuming data['JobTitle'] contains categorical data
job_title_counts = data['JobTitle'].value_counts().head()

# Plotting the pie chart using Matplotlib
plt.figure(figsize=(10, 8))
plt.pie(job_title_counts, labels=job_title_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Job Titles')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming data['JobTitle'] contains categorical data
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")
sns.countplot(y=data['JobTitle'], order=data['JobTitle'].value_counts().index[:10], palette="viridis")

plt.title('Top 10 Job Titles Distribution')
plt.xlabel('Count')
plt.ylabel('Job Title')
plt.show()
```

```python
data['BasePay'] = pd.to_numeric(data['BasePay'], errors='coerce')

# Filter numeric data
numeric_data = data[pd.notnull(data['BasePay'])]

# Plotting the bar graph
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")
sns.barplot(x='BasePay', y='JobTitle', data=numeric_data, errorbar=None, palette="viridis")

plt.title('Mean BasePay for Each Job Title')
plt.xlabel('Mean BasePay')
plt.ylabel('Job Title')
plt.show()
```
